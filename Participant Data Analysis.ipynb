```python
import numpy as np
import pandas as pd

# ignoring the warnings
import warnings
warnings.filterwarnings('ignore')

# We will display plots right inside Jupyter Notebook
%matplotlib inline
import matplotlib.pyplot as plt

# We will use the Seaborn library
import seaborn as sns
sns.set()

# Graphics in SVG format are more sharp and legible. Sometimes it can take longer to load.
# Use : %config InlineBackend.figure_format = 'png' in that case
%config InlineBackend.figure_format = 'svg'
```

```python
df = pd.read_csv('Participant_Data.csv')
df = df.drop(['Unnamed: 0'],axis=1)
```

```python
def convert_to_sec(time):
    minutes,sec = time.split(':')
    return int(minutes)*60 + int(sec)
```

```python
# Convert time to sec
df['TimeD1'] = df['TimeD1'].map(convert_to_sec)
df['TimeD2'] = df['TimeD2'].map(convert_to_sec)
```

```python
# Plot the distribution of timings between subjects analysis
plt.plot(df['TimeD1'],'bs')
plt.plot(df['TimeD2'],'rs')
plt.legend(['Without Assistant','With Assistant'])
plt.ylabel('Time (sec)')
plt.xlabel('Participant ID')
```

```python
# Mean timings for collective analysis
(df.mean()[1:3]).plot(kind='bar')
plt.xticks(range(2), ['Without Assistant','With Assistant'], rotation=0)
plt.ylabel('Time (sec)')
```
